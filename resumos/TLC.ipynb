## Teoria

#### Tendo uma m√©dia e uma vari√¢ncia ou desvio padr√£o, o teorema do limite central serve para calcular e mostrar as m√©dias amostrais, usando a seguinte f√≥rmula:

#### (segue uma distribui√ß√£o normal para um n grande o suficiente)

## Uso: plotando um gr√°fico com os valores de X e de Xbarra

```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
from scipy import stats


m√©dia = 8.2
var = pow(3.5, 2)
desvio_padr√£o = 3.5
n = 49

valor_do_scale = sqrt(var / n)

# criando uma lista com v√°rios valores para serem usados:
x = np.linspace(-5, 25, 1000)

# construindo a linha de X
f_x = stats.norm.pdf(x, loc=m√©dia, scale=desvio_padr√£o)

# construindo a linha de Xbarra
f_xBarra = stats.norm.pdf(x, loc=m√©dia, scale=valor_do_scale)


# plotando resultados
plt.figure(figsize=(16, 6))

plt.subplot(121)
plt.plot(x, f_x)
plt.xlabel('Valores de X')
plt.ylabel('Densidade')
plt.title(f'X ~ Normal, E(X)=8.2 e Var(X)={var}')

plt.subplot(122)
plt.plot(x, f_xBarra, color='orange')
plt.xlabel('Valores de Xbarra')
plt.xlim(5, 11)
plt.ylabel('Densidade')
plt.title(f'Xbarra ~ Normal, E(Xbarra)=8.2 e Var(Xbarra)={var}/49')


plt.show()
```

<h2>Resolvendo Problema </h2>

<h3>Assuma que mu=3.2 minutos. Encontre a probabilidade de que o tempo m√©dio de espera na fila seja:</h3>
<ul>
    <li>Menor do que 10 minutos</li>
    <li>Entre 8 e 10 minutos</li>
    <li>Maior do que 9 minutos</li>
</ul>

<h3>Sendo o tempo m√©dio uma vari√°vel qualquer com os seguintes par√¢metros:</h3>
<p>desvio padr√£o (sigma) = 3.5</p>

```python
mu = 8.2
n = 49
desvpad = 3.5
var = pow(desvpad, 2)

valor_do_scale = sqrt(var / n)

probability_maior10 = stats.norm.cdf(10, loc=mu, scale=valor_do_scale)
print(f'Menor do que 10 minutos: {probability_maior10:.5f}')

probability_entre8e10 = stats.norm.cdf(10, loc=mu, scale=valor_do_scale) - stats.norm.cdf(8, loc=mu, scale=valor_do_scale)
print(f'Entre 8 e 10 minutos: {probability_entre8e10:.5f}')

probability_maior9 = stats.norm.sf(9, loc=mu, scale=valor_do_scale)
print(f'Maior do que 9 minutos: {probability_maior9:.5f}')
```
